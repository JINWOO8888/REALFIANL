```python
import pandas as pd

products = pd.read_csv('./dataset/product_concat_v2.csv') 
```

```python
import re

print('p 태그 개수:' , len(products[products['CATALOG_DESC'].str.contains('<p style')]['CATALOG_DESC']))
print('th 태그 개수:' , len(products[products['CATALOG_DESC'].str.contains('<th')]['CATALOG_DESC']))
```

```python
products['CATALOG_DESC'] = products['CATALOG_DESC'].str.replace('\r',' ').str.replace('\\',' ').str.replace('\n',' ')
```

```python
# <부터 '까지의 패턴 정의
pattern = r'<[^>]*'

# 정규표현식을 사용하여 지우고 싶은 부분 삭제
products['CATALOG_DESC'] = products['CATALOG_DESC'].apply(lambda x: re.sub(pattern, '', x))
```

```python
print('p 태그 개수:' , len(products[products['CATALOG_DESC'].str.contains('<p style')]['CATALOG_DESC']))
print('th 태그 개수:' , len(products[products['CATALOG_DESC'].str.contains('<th')]['CATALOG_DESC']))
```
